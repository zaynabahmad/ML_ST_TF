```python
#to generate tensors use tf.variable
tensor1 = tf.Variable(["new", "one ", "vaiable "],tf.string)
tensor2 = tf.Variable(2,tf.int32)
tensor3 = tf.Variable([[2,3,5],[2,3,5]],tf.int32)
```

```python
#to get the rank of the tensors which is the order of its vector use
#tf.rank
print(tf.rank(tensor1))
print(tf.rank(tensor2))
print(tf.rank(tensor3))
#and to get the shape of the tensor which how many numbers in the vector use tf.shape
print("the shape of the tensor one is "+str(tf.shape(tensor1)))
```
